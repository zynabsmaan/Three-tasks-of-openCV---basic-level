```python
# import needed packages 
import torch
from torch.nn import functional
import numpy as np
import seaborn as sns
```

```python
# Define inputs  and outputs tensor  and needed functions 

def plot_function(x, y, function_name):
  sns.set_theme()
  p = sns.lineplot(x=torch.flatten(x), y=torch.flatten(y))
  p.set_xlabel("x")
  p.set_ylabel("y")
  p.set_title(function_name)


# Define and initalize inputs 
x = torch.tensor(np.reshape(np.arange(-10, 10, .5), (-1, 1)), dtype=torch.float)

# Define and initalize weights 
weights = torch.tensor([1.], dtype=torch.float)


print(f"The input is: {x}")
print(f"The weights is: {weights}")
```

## **Q1**

**Linear Units**

```python
y = functional.linear(x, weight=weights, bias=None)
plot_function(x, y, "Linear")
```

**Sigmoid Units**

```python
y = functional.sigmoid(x)
plot_function(x, y, "Sigmoid")
```

**Hyperbolic Tangent Units**

```python
y = functional.tanh(x)
plot_function(x, y, "Hyperbolic Tangent Units")
```

**Rectified Linear Units (ReLU)**

```python
y = functional.relu(x)
plot_function(x, y, "Rectified Linear Units (ReLU)")
```

```python
# Define and initialize input 
x = torch.tensor([[1, 0]], dtype=torch.float)

# define and initalize hidden weights 
weights_1 = torch.tensor([[.8, .4, .3],[.2, .9, .5]], dtype=torch.float)
weights_2 = torch.tensor([[.3],[.5], [.9]], dtype=torch.float)

# Define and initalize actual output
actual_output = torch.tensor([[1]], dtype=torch.float, requires_grad=True)
```

```python
# compute hidden layers 
hidden_1 = torch.sigmoid(torch.matmul(x, weights_1))

# compute output
output = torch.matmul(hidden_1, weights_2)
output
```

```python
# comoute MSE error
mse = (actual_output - output).pow(2).sqrt()
print(f"The MSE is {mse.item()}")
```
